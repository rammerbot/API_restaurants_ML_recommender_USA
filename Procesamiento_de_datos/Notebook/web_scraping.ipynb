# Webscraping de PIB Anual y PIB Per Capita

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC

estados=['alaska','arizona','arkansas','california','carolina-del-norte','carolina-del-sur','colorado','connecticut','dakota-del-norte','dakota-del-sur','delaware','florida','georgia','hawai','idaho','illinois','indiana','iowa','kansas','kentucky','luisiana','maine','maryland','massachusetts','michigan','minnesota','misisipi','misuri','montana','nebraska','nevada','nueva-jersey','nueva-york','nuevo-hampshire','nuevo-mexico','ohio','oklahoma','oregon','pensilvania','rhode-island','tennessee','texas','utah','vermont','virginia','virginia-occidental','washington','wyoming']

for estado in estados:
    chrome_driver_path = '../chromedriver-win64/chromedriver.exe'  


    chrome_options = Options()
    chrome_options.headless = True  # Ejecuta el navegador en modo sin cabeza (sin GUI)

    # Configura el servicio de ChromeDriver
    service = Service(chrome_driver_path)

    # Inicializa el navegador Chrome
    driver = webdriver.Chrome(service=service, options=chrome_options)

    # Navega a la página web
    driver.get('https://datosmacro.expansion.com/pib/usa-estados/'+estado)
    # Esperar a que la tabla esté presente en el DOM
    try:
        # Esperar hasta que el botón sea clickeable
        wait = WebDriverWait(driver, 2)
        accept_button = wait.until(EC.element_to_be_clickable((By.ID, 'ue-accept-notice-button')))
        accept_button.click()
        print("Botón 'Aceptar y continuar' clickeado exitosamente.")
    except Exception as e:
        print(f"No se pudo hacer clic en el botón: {e}")

    try:
        driver.implicitly_wait(2)
        button = driver.find_element(By.ID, 'btnDi2')
        button.click()

        try:
            # Espera a que el modal esté presente
            modal = WebDriverWait(driver, 2).until(
                EC.presence_of_element_located((By.CLASS_NAME, 'modal-dialog'))
            )
            
            # Espera a que el botón de cierre esté presente y sea clickeable
            close_button = WebDriverWait(driver, 2).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, 'button.close[data-dismiss="modal"]'))
            )
            
            close_button.click()
            print("Botón de cierre clickeado exitosamente")
        except Exception as e:
            print(f"Error al intentar hacer clic en el botón del div modal-dialog: {e}")

        div_raiz = driver.find_elements(By.CLASS_NAME, 'content')   
    
        # Abre el archivo para escribir
        with open('../Data-externa/pib_'+estado+'.txt', 'w', newline='',encoding='utf-8') as file:
            for div in div_raiz:
                # Divide el texto del div en líneas
                lines = div.text.split('\n')
                for line in lines:
                    # Validar para no escribir basura
                    if 'PIB tasa de crecimiento a precios constantes ' not in line:
                        # Escribe cada línea en el archivo
                        file.write(line + '\n')
                    else :
                        break
            
    except Exception as e:
        print(f"Error: {e}")
        driver.quit()
```

# Webscraping de densidad de población

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC

estados=['alaska','arizona','arkansas','carolina-del-norte','carolina-del-sur','colorado','connecticut','dakota-del-norte','dakota-del-sur','delaware','florida','georgia','hawai','idaho','illinois','indiana','iowa','kansas','kentucky','luisiana','maine','maryland','massachusetts','michigan','minnesota','misisipi','misuri','montana','nebraska','nevada','nueva-jersey','nueva-york','nuevo-hampshire','nuevo-mexico','ohio','oklahoma','oregon','pensilvania','rhode-island','tennessee','texas','utah','vermont','virginia','virginia-occidental','washington','wyoming']

for estado in estados:
    chrome_driver_path = '../chromedriver-win64/chromedriver.exe'  

    # Configura las opciones de Chrome
    chrome_options = Options()
    chrome_options.headless = True  # Ejecuta el navegador en modo sin cabeza (sin GUI)

    # Configura el servicio de ChromeDriver
    service = Service(chrome_driver_path)

    # Inicializa el navegador Chrome
    driver = webdriver.Chrome(service=service, options=chrome_options)

    # Navega a la página web
    driver.get('https://datosmacro.expansion.com/demografia/poblacion/usa-estados/'+estado)


    try:
        # Esperar hasta que el botón sea clickeable
        wait = WebDriverWait(driver, 2)
        accept_button = wait.until(EC.element_to_be_clickable((By.ID, 'ue-accept-notice-button')))
        accept_button.click()
        print("Botón 'Aceptar y continuar' clickeado exitosamente.")
    except Exception as e:
        print(f"No se pudo hacer clic en el botón: {e}")

    try:
        div_raiz = driver.find_elements(By.CLASS_NAME, 'content')   
        print(div_raiz[0].text)
        head=0
        # Abre el archivo para escribir
        with open('Data-externa/densidad_'+estado+'.txt', 'w', newline='',encoding='utf-8') as file:
            for div in div_raiz:
                # Divide el texto del div en líneas
                lines = div.text.split('\n')
                for line in lines:
                    if head<2:
                        if estado+' - población' in line.lower():
                            head+=1
                        file.write(line + '\n')
                    else :
                         break

    except Exception as e:
        print(f"Error: {e}")
    driver.quit()
```
