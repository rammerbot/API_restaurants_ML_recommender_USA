Manipulación de columnas y datos según lo requerido

```python
# Definimos las columnas que queremos conservar para calcular total_valoraciones
valoraciones_a_sumar = ['useful', 'funny', 'cool', 'compliment_hot', 
'compliment_more', 'compliment_profile', 'compliment_cute', 
'compliment_list', 'compliment_note', 'compliment_plain', 
'compliment_cool', 'compliment_funny', 'compliment_writer', 
'compliment_photos']
```

```python
# Sumamos las columnas seleccionadas para obtener total_valoraciones
df_user['total_valoraciones_positivas'] = df_user[valoraciones_a_sumar].sum(axis=1)
```

```python
# Selecciona solo las columnas deseadas
columnas_mantenidas = ['user_id', 'review_count', 'average_stars', 'total_valoraciones_positivas', 'fans']

# Filtra el DataFrame para conservar solo las columnas especificadas
df_user = df_user[columnas_mantenidas]

# Verifica el DataFrame resultante
print(df_user.head())
```

Nulos, duplicados, faltantes

```python
warnings.filterwarnings('ignore')
```

```python
# Comprobación de que no hay valores nulos
df_user.isnull().sum()
```

```python
# Identificar duplicados
duplicados = df_user[df_user.duplicated()]

# Imprimir el número de duplicados encontrados
print(f"Número de filas duplicadas: {duplicados.shape[0]}")
```

```python
# Eliminar filas duplicadas
df_user = df_user.drop_duplicates()
```

```python
# Redondear la columna 'average_stars' a un decimal
df_user['average_stars'] = df_user['average_stars'].round(1)
```
